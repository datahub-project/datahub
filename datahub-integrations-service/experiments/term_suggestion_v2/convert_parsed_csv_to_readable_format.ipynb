```python
import pandas as pd
import pickle
```

```python
with open("parsed_response.pkl", 'rb') as f:
    parsed_response = pickle.load(f)
    f.close()
```

```python
parsed_response
```

```python
column_terms_dict = {}
table_urns=[]
instances = []
columns = []
column_terms = []
# column_terms = {}
for table, instance, table_terms, all_column_terms in parsed_response:
    # table_urns.append(table)
    # instances.append(instance)
    for column, terms in all_column_terms.items():
        table_urns.append(table)
        instances.append(instance)
        columns.append(column)
        column_terms.append([(term.name, term.confidence_score, term.reasoning, term.is_fake) for term in terms])
            
```

```python
df = pd.DataFrame({"urn": table_urns, "instance": instances, "column": columns, "assigned_terms": column_terms})
```

```python
df['high_conf_terms'] = df.assigned_terms.apply(lambda x: [(term[0], term[1], term[2]) for term in x if term[1]>=9 and not term[3]])
```

```python
df.to_csv("assigned_terms.csv")
```
