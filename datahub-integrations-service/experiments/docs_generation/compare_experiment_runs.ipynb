```python
# Claude 3 vs Claude 3.5
# run_name1 = "carefree-skink-867"  # claude 3.5
# run_name2 = "tasteful-penguin-228"  # claude 3 haiku


# Benchamrk vs new run (links, etc)
run_name1 = "tasteful-penguin-228"  # old benchmark
run_name2 = "learned-rat-959"  # linkify + minor changes

suffixes = ("_benchmark", "_new")
```

```python
import pandas as pd
from eval_common import METRIC_NAMES
from mlflow_common import get_ai_eval_result_or_none

result1 = get_ai_eval_result_or_none(run_name1)
result2 = get_ai_eval_result_or_none(run_name2)
columns = ["urn", "description", "overall_score/score", "overall_score/justification"]
for metric_name in METRIC_NAMES:
    columns.append(f"{metric_name}/score")
    columns.append(f"{metric_name}/justification")

result1 = result1[columns]
result2 = result2[columns]


merged_result = pd.merge(result1, result2, on="urn", suffixes=suffixes)
```

```python
metric_name = "has_usage_tips"
degraded_metric_df = merged_result[
    (
        merged_result[f"{metric_name}/score_benchmark"]
        > merged_result[f"{metric_name}/score_new"]
    )
]

pd.options.display.max_colwidth = None
display(
    degraded_metric_df[
        [
            "urn",
            f"{metric_name}/score_benchmark",
            f"{metric_name}/score_new",
            f"{metric_name}/justification_benchmark",
            f"{metric_name}/justification_new",
        ]
    ]
)
```

```python
improved_metric_df = merged_result[
    (
        merged_result[f"{metric_name}/score_benchmark"]
        < merged_result[f"{metric_name}/score_new"]
    )
]

pd.options.display.max_colwidth = None
display(
    improved_metric_df[
        [
            "urn",
            f"{metric_name}/score_benchmark",
            f"{metric_name}/score_new",
            f"{metric_name}/justification_benchmark",
            f"{metric_name}/justification_new",
        ]
    ]
)
```
